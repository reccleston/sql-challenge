```python
import pandas as pd
```

```python
salaries = pd.read_csv('salaries.csv')
employees = pd.read_csv('employees.csv')
dept_manager = pd.read_csv('dept_manager.csv')
dept_emp = pd.read_csv('dept_emp.csv')
dept = pd.read_csv('departments.csv')
titles = pd.read_csv('titles.csv')
```

```python
raw_tables = [salaries, employees, dept_manager, dept_emp, dept, titles]
```

```python
def isUnique(df):
    cols = df.columns
    for i,col in enumerate(cols):
        print(f"proposed pk: {col}, data: {', '.join(cols[i+1:])} ")
        return len(df.loc[:,f'{col}']) == len(df.loc[:,f'{col}'].unique()) 
```

```python
for table in raw_tables:
    print(isUnique(table))
```

```python
def getColType(df):
    return df.dtypes
```

```python
for table in raw_tables:
    print(getColType(table), '\n')
```
